```python
# removing un-necessary cols : 

df.drop(columns=['Serial No.'],inplace=True)
```

```python
df.head()
```

```python
# separating input rows and output rows : 
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# checking any duplicate rows : 
df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
# scaling the values , so that the NN can train properly : 

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation='relu',input_dim=7))
model.add(Dense(7,activation='relu')) # hidden layer 1 
model.add(Dense(7,activation='relu')) # hidden layer 2
model.add(Dense(7,activation='relu')) # hidden layer 3
model.add(Dense(1,activation='linear')) # for every Regression Problem the activation function will be 'linear'
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
history = model.fit(X_train_scaled,y_train,epochs=50,validation_split=0.2)
```

```python
y_pred=model.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
